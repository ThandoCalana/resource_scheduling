```python
# we drop Field7 and planned Exam date feature since they have more than 80% of the missing values.
df.drop(['Field7', 'Planned Exam Date'], axis=1, inplace=True)
```

```python
partner_zero_rows = df[df['Partner'] == 2]
display(partner_zero_rows)
```

```python
df['Cost ($)'] = df['Cost ($)'].fillna(0)
print("Missing values in 'Cost ($)' after filling with 0:")
print(df['Cost ($)'].isnull().sum())
```

```python
# Calculate 'validityNow' as the difference between 'Expiration Date' and the current date
current_date = pd.Timestamp.now()
df['validityNow'] = df['Expiration Date'] - current_date

display(df.head())
```
